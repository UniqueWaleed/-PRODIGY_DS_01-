Create a bar chart or histogram to visualize the distribution of a categorical or continuous variable.

**BAR CHART**

```python
import pandas as pd
import matplotlib.pyplot as plt

modified_plotdata = pd.DataFrame({
    "2022": [85, 78, 88, 90],
    "2023": [88, 82, 92, 91],
    "2024": [90, 86, 94, 93]
}, index=["Player A", "Player B", "Player C", "Player D"])


ax = modified_plotdata.plot(kind='bar', stacked=True, figsize=(15, 8))


ax.set_title("Footballers' Performance Ratings Over Time")
ax.set_xlabel("Footballer")
ax.set_ylabel("Ratings")
ax.legend(title="Year")

plt.show()
```

**HISTOGRAM**

```python
import pandas as pd
import matplotlib.pyplot as plt


modified_plotdata = pd.DataFrame({
    "2022": [85, 78, 88, 90],
    "2023": [88, 82, 92, 91],
    "2024": [90, 86, 94, 93]
}, index=["Player A", "Player B", "Player C", "Player D"])


ratings_2022 = modified_plotdata["2022"]


plt.figure(figsize=(10, 6))
plt.hist(ratings_2022, bins=10, color='skyblue', edgecolor='black')
plt.title("Histogram of Footballers' Ratings in 2022")
plt.xlabel("Ratings")
plt.ylabel("Frequency")

plt.show()
```
